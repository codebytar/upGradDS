### Calculate the Average Growth of States

#### Part 1 A
#### Using Data 1-A

Import GDP datasets into a pandas dataframe and import all other necessary python libraries


```python
#import pandas, matplotlib, numpy and seaborn libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#read the csv data into a pandas dataframe

df = pd.read_csv(r"C:\Users\Tariq Hussain\Desktop\UpGrad\GDP Analysis\GDP_Data.csv")
```


**Remove the rows: '(% Growth over the previous year)' and 'GSDP - CURRENT PRICES (` in Crore)' for the year 2016-17.**


```python
#Remove the rows that are no longer required

df = df.drop(df.index[[5]])
df = df.drop(df.index[[9]])

df
```


**Calculate the average growth of states over the duration 2013-14, 2014-15 and 2015-16**


**Clean and Transform datasets**


```python
#create a new dataframe with only Growth percentage values of all states and union territories

df_avg_growth = df[6:]

#drop the column 'Items Description'

df_avg_growth = df_avg_growth.drop(df.columns[[0]], axis = 1)

#Transpose the datasets

df_avg_growth = df_avg_growth.transpose()

#Rename Column Names

df_avg_growth.columns = ['2013_14','2014_15','2015_16']

#Remove Rows
df_avg_growth = df_avg_growth[1:]

#Rename the axis
df_avg_growth = df_avg_growth.rename_axis('State')


df_avg_growth
```


**Calculate the average growth rate for all Indian States in a new column within the dataframe**


```python
#Replace NaN values with 0

df_avg_growth = df_avg_growth.fillna(0)

#Calculate mean values in a new column for each state

meancol = df_avg_growth.loc[:,"2013_14":"2015_16"]

df_avg_growth['Average_Growth'] = round(meancol.mean(axis=1),2)

#Sort datasets in decending order

df_avg_growth = df_avg_growth.sort_values(by = 'Average_Growth', ascending = False)


df_avg_growth
```


**Plot the datasets**


```python
#Set the chart background and size

sns.set(style ="whitegrid")
plt.figure(figsize=(15,10))

#Plot the datasets

growth_plot = sns.barplot(x='Average_Growth',y =df_avg_growth.index, data = df_avg_growth, color='b')

#Add Title and Rename the labels

plt.title('Average Growth Percentage of Indian State from 2013 to 2016', fontsize =16)
plt.ylabel('States', fontsize = 14)
plt.xlabel('Average Growth Percentage', fontsize = 14)

plt.show()
```



**Which states have been growing consistently fast, and which ones have been struggling?**

#### Calculate and plot per capita GDP for all States
#### Calculate and plot percentage contribution by Primary, Secondary and Tertiary sectors to the overall GDP of a State

```python
#Calculate per capita GDP for all states
#Calculate contribution percentages for Primary, Secondary and Tertiary

perCapitaGDP =  df_state_gdp_per_capita['GDP']/df_state_gdp_per_capita['Population']
primaryPercentage = df_state_gdp_per_capita['Primary']/df_state_gdp_per_capita['GDP']
secondaryPercentage = df_state_gdp_per_capita['Secondary']/df_state_gdp_per_capita['GDP']
tertiaryPercentage = df_state_gdp_per_capita['Tertiary']/df_state_gdp_per_capita['GDP']

df_state_gdp_per_capita['Per Capita GDP'] = round(perCapitaGDP,2)
df_state_gdp_per_capita['Primary Percentage'] = round(primaryPercentage,2)
df_state_gdp_per_capita['Secondary Percentage'] = round(secondaryPercentage,2)
df_state_gdp_per_capita['Tertiary Percentage'] = round(tertiaryPercentage,2)

#Sort the dataframe based on per capital GDP

df_state_gdp_per_capita = df_state_gdp_per_capita.sort_values(by ='Per Capita GDP' , ascending = False).reset_index()

#Drop unwanted columns

df_state_gdp_per_capita = df_state_gdp_per_capita.drop(df_state_gdp_per_capita.columns[[0]], axis = 1)

df_state_gdp_per_capita
```
